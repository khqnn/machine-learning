#### Choose which columns to use in clustering

* pick with similar ranges

```python
X = df.loc[:, ('artist_top_genre','popularity','danceability','acousticness','loudness','energy')]

y = df['artist_top_genre']
```

#### Encode artist_top_genre to numeric

```python
le = LabelEncoder()
X['artist_top_genre'] = le.fit_transform(X['artist_top_genre'])
y = le.transform(y)
```

#### Use K-means clustring algorithm

```python
nclusters = 3 
seed = 0

km = KMeans(n_clusters=nclusters, random_state=seed)
km.fit(X)
```

#### Predict cluster for each datapoint

```python
y_cluster_kmeans = km.predict(X)
y_cluster_kmeans
```

#### calculate silhouette score

```python
score = metrics.silhouette_score(X, y_cluster_kmeans)
score
```

#### Try with different number of clusters

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

#### use elbow method to choose good number of clusters

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Display the cluster

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
labels = kmeans.predict(X)
plt.scatter(df['popularity'],df['danceability'],c = labels)
plt.xlabel('popularity')
plt.ylabel('danceability')
plt.show()
```

#### Calculate the model's accuracy

```python
labels = kmeans.labels_

correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```
